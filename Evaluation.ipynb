```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

def load_txt_files(base_folder, file_name, start=0.1, end=1.0, step=0.1):
    """
    Iteratively reads `.txt` files from folders with names in the format `x_y`
    (e.g., `0.1_0.1`, `0.1_0.2`, ...) and loads their contents as lists.

    Args:
        base_folder (str): Path to the base directory containing the folders.
        start (float): Starting value for x and y.
        end (float): Ending value (exclusive) for x and y.
        step (float): Step size for x and y.

    Returns:
        dict: A dictionary where keys are folder names (e.g., `0.1_0.1`) 
              and values are lists of floats from the `.txt` files.
    """
    data = {}  # To store the results
    x = start

    while x < end:
        y = start
        while y < end:
            folder_name = f"{x:.1f}_{y:.1f}"  # Generate folder name
            folder_path = os.path.join(base_folder, folder_name)
            txt_file_path = os.path.join(folder_path, file_name)  # Target file path
            
            if os.path.exists(txt_file_path):  # Check if the file exists
                # Read the `.txt` file and store its contents as a list of floats
                with open(txt_file_path, "r") as f:
                    data[folder_name] = [float(line.strip()) for line in f]
                print(f"Loaded data from: {txt_file_path}")
            else:
                print(f"File not found: {txt_file_path}")
            
            y += step
        x += step

    return data

# Example usage
base_folder = "./Trajectories"  # Base directory containing the folders
file_name = "Lower_bound_list.txt"
lower_bound = load_txt_files(base_folder,file_name, start=0.1, end=1.0, step=0.1)
return_file = "return_list.txt"
return_list = load_txt_files(base_folder,return_file, start=0.1, end=1.0, step=0.1)

# Display a preview of the loaded data
# for folder, values in data.items():
#     print(f"Folder: {folder}, First 5 Values: {values[:5]}")
```

```python
return_mean = []
return_var = []
for folder, values in return_list.items():
    return_mean.append(np.mean(values))
    return_var.append(np.var(values))

con_clb = []

for folder,values in lower_bound.items():
    con_clb.append(np.mean(values[-5:]))

    
```

```python
def plot_results(data):
    """
    Plots all results on the same plot.

    Args:
        data (dict): A dictionary where keys are folder names and values are lists of results.
    """
    plt.figure(figsize=(14, 6))
    for folder_name, values in data.items():
        plt.plot(values, label=folder_name)  # Plot each folder's data with its name as the label
    
    #plt.title("Results from Different Folders")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.legend(loc="lower center", bbox_to_anchor=(1, 1), title="Dev_Prob",ncol=9)
    plt.tight_layout()
    plt.show()
```

```python
plot_results(lower_bound)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def draw_heatmap_from_dict(mean_rewards, start=0.1, end=1, step=0.1):
    """
    Draws a heatmap of mean rewards directly from a dictionary.

    Args:
        mean_rewards (dict): A dictionary where keys are folder names (e.g., "0.1_0.2")
                             and values are the mean rewards for those folders.
        start (float): Starting value for the folder names (e.g., 0.1).
        end (float): Ending value (exclusive) for the folder names (e.g., 1.0).
        step (float): Step size for the folder names (e.g., 0.1).
    """
    # Prepare axis labels
    y_values = np.arange(start, end, step)  # First part (Y-axis)
    x_values = np.arange(start, end, step)  # Second part (X-axis)
    
    # Create an empty 2D array for the heatmap
    heatmap_data = np.zeros((len(y_values), len(x_values)))

    # Populate the heatmap with mean rewards from the dictionary
    for i, y in enumerate(y_values):
        for j, x in enumerate(x_values):
            folder_name = f"{y:.1f}_{x:.1f}"  # Generate folder name
            heatmap_data[i, j] = mean_rewards.get(folder_name, np.nan)  # Use NaN if folder is missing

    # Convert the 2D array into a DataFrame for better labeling
    heatmap_df = pd.DataFrame(
        heatmap_data,
        index=[f"{y:.1f}" for y in y_values],  # Y-axis labels
        columns=[f"{x:.1f}" for x in x_values]  # X-axis labels
    )

    # Plot the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        heatmap_df,
        annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'CLB'},
        linewidths=0.5, linecolor="black"
    )
    plt.title("Heatmap of CLB")
    plt.xlabel("Second Part (X-Axis)")
    plt.ylabel("First Part (Y-Axis)")
    plt.show()
```
